Step 1: Loaded the Iris DATASET and did an overview

reffred: https://scikit-learn.org/1.5/auto_examples/datasets/plot_iris_dataset.html

```python
#imported all the libraries needed
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

print("First few rows of the Iris dataset:")
print(iris.head())

print("\nDataset Info:")
print(iris.info())
```

Step 2: Performed Simple Random Sampling:

Took 30 samples from the data set

```python
sample_random = iris.sample(n=30, random_state=42)

print("\nRandom Sample of 30 Observations:")
print(sample_random)
```

Step 3: Sample Mean Distribution Analysis :
Calculated the mean for the sepal length , random sampling was performed for 100 sample range
used histogram to show the distribution of the selected samples as it highlights the spread based on the sample mean

```python
sample_means = [iris['sepal_length'].sample(n=30, random_state=i).mean() for i in range(100)]

plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=15, color='maroon', alpha=0.7, edgecolor='black')
plt.title('Distribution of Sample Means (Sepal Length)')
plt.xlabel('Sample Mean (Sepal Length)')
plt.ylabel('Frequency')
plt.show()
```

Step 4: Systematic Sampling :

Took 20% of the dataset, use a custom function to perform systematic sampling.
Kept the orginal data set plot wit the systematic sampling to compare the distribution of teh sepal length.

```python
def systematic_sampling(data, step):

    indices = np.arange(0, len(data), step)
    return data.iloc[indices]

step = len(iris) // (len(iris) // 5)
sample_systematic = systematic_sampling(iris, step)

print("\nSystematic Sample (20% of Dataset):")
print(sample_systematic)


plt.figure(figsize=(14, 6))

# Original dataset
plt.subplot(1, 2, 1)
sns.histplot(iris['sepal_length'], kde=True, bins=15, color='blue')
plt.title('Original Dataset (Sepal Length)')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')

# Systematic sampled dataset
plt.subplot(1, 2, 2)
sns.histplot(sample_systematic['sepal_length'], kde=True, bins=15, color='purple')
plt.title('Systematic Sample (20%) - Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Step 1: Importing Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Step 2: Loaded data set using Real time data from github resource

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data = pd.read_csv(url)

# Display the first few rows
print(data.head())
```

Step 3 : Performed descriptive analysis for the dataset taken.

          1. Measures of Central Tnedency : Performed Mean, Median and Mode - in order to determine centre of the data.
          
          2. Measures of Dispersion : Performed Variance, Standard Deviation and Range - in order to determine how the data spread is.

```python
# Descriptive statistics
print("Dataset Overview:")
print(data.info())

print("\nDescriptive Statistics:")
print(data.describe())

# Central tendency and dispersion for specific columns

target_column = "Fare"
print("\nCentral Tendency and Dispersion for 'Fare':")
print(f"Mean: {data[target_column].mean()}")
print(f"Median: {data[target_column].median()}")
print(f"Mode: {data[target_column].mode().iloc[0]}")
print(f"Variance: {data[target_column].var()}")
print(f"Standard Deviation: {data[target_column].std()}")
print(f"Range: {data[target_column].max() - data[target_column].min()}")
```

Step 4 : Visualization

1. Mapping Ticket Trends.

I analyzed the fare distribution to understand ticket pricing trends and patterns. The visualization revealed that most passengers paid lower fares, while a few paid much higher, likely representing first-class passengers.



```python
# Histogram and KDE for 'Fare'

plt.figure(figsize=(10, 6))
sns.histplot(data['Fare'], kde=True, bins=30, color='blue')
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

2. Age Influence

To explore the relationship between age, fare, and survival, I created a scatter plot. It showed that younger passengers and those who paid higher fares were more likely to survive, hinting at the influence of socio-economic factors.

```python
# Scatter plot: Fare vs. Age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=data, hue='Survived', palette='coolwarm')
plt.title('Age vs. Fare (Colored by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

3. Survivors vs Non Survivors

I plotted a bar chart to compare survivors with non-survivors. It highlighted that fewer people survived, with survival rates falling below 50%, giving a quick overview of the tragedy's impact.

```python
# Bar plot: Survival count
plt.figure(figsize=(8, 5))
sns.countplot(x='Survived', data=data, palette='viridis')
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```
